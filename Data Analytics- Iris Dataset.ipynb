```python
# Importing all Desired library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Read the csv File.
iris=pd.read_csv('iris.csv')
```

```python
# Read the Top 5 Datapoints of the Dataset.
iris.head()
```

```python
# To Determine the Shape of the Dataset.
iris.shape
```

```python
# To Determine the Dimension of the Dataset.
iris.ndim
```

```python
# To Determine the size of the Dataset.
iris.size
```

```python
# To Describe the dataset. (like to find the maximum value , minimum value, mean, standard deviationetc.)
iris.describe()
```

```python
# To Determine the median of the dataset
iris.median()
```

```python
# To Determine only the mean of the Dataset
iris.mean()
```

As Here we Analysis that the mean is approxiatly same as the median of the dataset so now we conclude that the
Data is in Normal Distribution so the bell curve of this data is symetric i.e there is no skweness.

```python
# To Visualise the sepal length's mean,minimum value,maximum value,25%,75% from the dataset.
sns.boxplot(data=iris['Sepal.Length'])
```

```python
# To Determine the Variation in the Dataset
iris.var()
```

As here the variation is very low expect the petal length so we here conclude that the there is no very variation in the dataset.

```python
# To Determine the value counts of the species.
iris.Species.value_counts()
```

Here we conclude that all the 3 species comes equally.

```python
# To Visualise the value counts of the species.
plt.figure(figsize=(5,5))
myex=[0.1,0,0]
plt.pie(iris.Species.value_counts(),labels=iris.Species.value_counts().keys(),startangle=90,autopct='%0.1f%%',explode=myex)
plt.show()
```

```python
# To Get the Information of the dataset.
iris.info()
```

```python
# To Determine the petal length and petal width of those data who's sepal length is less than 5
iris[iris['Sepal.Length']<5][['Petal.Length','Petal.Width']][0:5]
```

```python
# To Determine the Null values.
iris.isna().head()
```

```python
sns.displot(iris['Sepal.Length'],bins=150,kde=True,color='k')
```

```python
# To Determine how many null values a dataset have
iris.isnull().sum()
```

```python
# To visualise how much the data is scattered.
plt.figure(figsize=(5,5))
sns.scatterplot(x='Sepal.Length',y='Petal.Length',color='hotpink',data=iris)
plt.title('Scatter Plot')
plt.grid()
plt.show()
```
